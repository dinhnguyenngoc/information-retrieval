```python
from pyspark.sql import SparkSession
```

```python
# Khởi tạo SparkSession
spark = SparkSession.builder \
    .appName("Read Text Example") \
    .getOrCreate()
```

```python
# Read a text file and convert each line to a Row
textFile = spark.read.text("README.md")
```

```python
# Count the number of rows in this DataFrame
print(textFile.count())
# Print the first row in this DataFrame
textFile.first()
```

```python
# Count how many lines contain "Spark"
# linesWithSpark = textFile.filter(textFile.value.contains("Spark"))
# linesWithSpark.count()
textFile.filter(textFile.value.contains("Spark")).count()
```

### More on Dataset Operations

```python
from pyspark.sql import functions as sf
textFile.select(sf.size(sf.split(textFile.value, "\s+")).name("numWords")).agg(sf.max(sf.col("numWords"))).collect()
```

```python
wordCounts = textFile.select(sf.explode(sf.split(textFile.value, "\s+")).alias("word")).groupBy("word").count()
wordCounts.collect()
```
