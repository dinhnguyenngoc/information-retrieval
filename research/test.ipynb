```python
from pyspark.sql import SparkSession
```

```python
# Khởi tạo SparkSession
spark = SparkSession.builder \
    .appName("Read CSV Example") \
    .getOrCreate()
```

```python
# Đọc file CSV
csvFile = spark.read.csv("posts.csv", header=True, inferSchema=True)
```

```python
# Hiển thị một số dòng dữ liệu
csvFile.show()
```
